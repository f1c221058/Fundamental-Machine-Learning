```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

Memisahkan dataset menjadi:

    Input (x): Ukuran rumah dalam meter persegi
    Output (y): Harga rumah dalam juta rupiah

Kemudian, dataset juga dibagi menjadi data training dan testing menggunakan metode train_test_split.

```python
data = {
    'Ukuran': [50, 60, 70, 80, 90, 100, 110, 120, 130, 140, 150],
    'Harga': [150, 180, 210, 240, 270, 300, 330, 360, 390, 420, 450]
}
df = pd.DataFrame(data)
```

```python
X = df[['Ukuran']]
y = df['Harga']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
X_test
```

Membuat object bernama model berdasarkan Class Linear Regression. Kemudian, kita latih modelnya dengan data testing (X dan y testing)

```python

model = LinearRegression()
model.fit(X_train, y_train)
```

lanjut dengan melakukan prediksi nilai berdasarkan data testing (X_test) dan hasilnya kita simpan dalam y_pred

```python
y_pred = model.predict(X_test)
```

Kode di atas akan menghasilkan grafik berikut :

```python
plt.scatter(X_test, y_test, color='blue', label='Data Sebenarnya')
plt.plot(X_test, y_pred, color='red', label='Model Regresi Linear')
plt.xlabel('Ukuran Rumah (m)')
plt.ylabel('Harga Rumah (juta rupiah)')
plt.title('Prediksi Harga Rumah')
plt.legend()
plt.show()
```

```python
print("Koefisien (Slope):",model.coef_[0])
print("Intercept:", model.intercept_)
```

## Model Regresi Linear

Model regresi linear berbentuk umum sebagai berikut:

\[
y = \beta_0 + \beta_1 x
\]

Di mana:
- \( y \) adalah variabel **target/output** (misalnya: harga)
- \( x \) adalah variabel **fitur/input** (misalnya: ukuran)
- \( \beta_0 \) adalah **intercept** (titik potong dengan sumbu y)
- \( \beta_1 \) adalah **koefisien regresi (slope)**

### Dengan Nilai yang Diberikan:

- Koefisien/Slope: \( \beta_1 = 2.9999999999999996 \)
- Intercept: \( \beta_0 = 0.0 \)

### Maka bentuk modelnya adalah:

\[
y = 3x
\]

Karena \( 2.9999999999999996 \approx 3 \) dan intercept-nya 0, maka:
> Setiap kenaikan 1 unit \( x \) akan meningkatkan \( y \) sebesar 3 unit, **tanpa pergeseran vertikal**.
